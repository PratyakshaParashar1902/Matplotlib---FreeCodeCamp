### Loading Libraries

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

### Baisc Graph

```python
# 1,3,5 are X axis and 2,4,6 are Y
plt.plot([1,3,5],[2,4,6])
```

```python
# The text printed above doesn't appear when we use show()
x = [0,1,3,5]
y = [0,2,6,10]

# Resizing Graph
plt.figure(figsize=(5,3), dpi=100)
#  If printing graph use a higher dpi


# Can add font and size to any of the labels
# Can also use Hex code for colouring
# plt.plot(x,y, label='2x', color = 'red', linewidth=2, marker='.', linestyle='--',markersize=10,markeredgecolor='blue')
# Using Shorthand notation
# fnt = '[color][marker][line]'
plt.plot(x,y,'b.--', label='2x')

# Line-2
# Select interval we want to plot at
x2 = np.arange(0,4.5,.5)

# Plotting part of the graph
plt.plot(x2[:6],x2[:6]**2,'r' ,label='x^2')
# Plotting remaining of the graph
plt.plot(x2[5:],x2[5:]**2,'r' ,linestyle='--')


plt.title("Our first Graph", fontdict = {'fontname': 'Comic Sans MS', 'fontsize': 20})
plt.xlabel("X Axis")
plt.ylabel("Y Axis")

# Can use ticks for marking values in our axis 
# plt.xticks([0,1,2,3,4])
# plt.yticks([0,2,4,6,8,10])

# Legend to show label
plt.legend()

# Saving our graph
plt.savefig('MyGraph.png', dpi = 300)
plt.show()
```

### Bar Charts

```python
labels = ['A','B','C']
values = [1,4,2]

bars = plt.bar(labels, values)

# Hatching our bars

# bars[0].set_hatch('/')
# bars[0].set_hatch('o')
# bars[0].set_hatch('*')

# OR

patterns = ['/','o','*']
for bar in bars:
    bar.set_hatch(patterns.pop(0))

plt.figure(figsize=(6,4))

plt.show()
```

### Loading FIFA data

```python
fifa = pd.read_csv('fifa_data.csv')
fifa.head(5)
```

#### Histograms

```python
bins = [40,50,60,70,80,90,100]

plt.hist(fifa.Overall, bins = bins)

plt.xlabel('Number of Players')
plt.ylabel('Skill level')
plt.title('Distribution of Player skills in FIFA World Cup 2018')

plt.show()
```

#### PIE Charts

```python
left = fifa.loc[fifa['Preferred Foot'] == 'Left'].count()[0]
right = fifa.loc[fifa['Preferred Foot'] == 'Right'].count()[0]
   
plt.figure(figsize=(8,5))
    
labels = ['Left','Right']
colors = ['#abcdea','#adcadc']

plt.pie([left,right], labels = labels, colors=colors, autopct='%.2f %%')

plt.title('Foot Preference of FIFA Players')

plt.show()
```

#### PIE Charts with more functionalities

```python
plt.figure(figsize=(8,5),dpi = 100)

# For colours palet
plt.style.use('ggplot')

fifa.Weight = [int(x.strip('lbs')) if type(x)==str else x for x in fifa.Weight]

light = fifa.loc[fifa.Weight < 125].count()[0]
light_medium = fifa.loc[(fifa.Weight >= 125) & (fifa.Weight < 150)].count()[0]
medium = fifa.loc[(fifa.Weight >= 150) & (fifa.Weight < 175)].count()[0]
medium_heavy = fifa.loc[(fifa.Weight >= 175) & (fifa.Weight < 200)].count()[0]
heavy = fifa.loc[fifa.Weight >= 200].count()[0]

weights = [light,light_medium,medium,medium_heavy,heavy]
label = ['under 25','125-150','150-175','175-200','over 200']
explode = (.4,.2,0,0,.4)

plt.title('Weight of Professional Soccer Players (lbs)')

plt.pie(weights, labels=label, explode=explode, pctdistance=0.8,autopct='%.2f %%')

plt.show()
```

#### BOX and Whiskers Chart

```python
plt.figure(figsize=(5,8), dpi=100)

plt.style.use('default')

barcelona = fifa.loc[fifa.Club == 'FC Barcelona']['Overall']
madrid = fifa.loc[fifa.Club == 'Real Madrid']['Overall']
revs = fifa.loc[fifa.Club == 'New England Revolution']['Overall']

labels = ['FC Barcelona','Real Madrid','NE Revolution']

boxes = plt.boxplot([barcelona,madrid,revs], labels=labels,patch_artist = True, medianprops={'linewidth':2})

for box in boxes['boxes']:
#     Set Edge color
        box.set(color='#abdfad', linewidth=2)
    
#     Change fill color
        box.set(facecolor='#e0e0e0')
    
plt.title('Professional Football Team Comparision')
plt.ylabel("FIFA Overall rating")

plt.show()
 
```
